```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a complete graph K4
K5 = nx.complete_graph(5)

# Define node positions for a symmetric layout
pos = nx.spring_layout(K5, seed=42)

# Plot the graph
plt.figure(figsize=(6, 6))
nx.draw(
    K5,
    pos,
    with_labels=True,
    node_color="lightblue",
    edge_color="blue",
    node_size=800,
    font_size=12,
    font_color="black",
    linewidths=1,
)
plt.title("Complete Graph K5 with Blue Edges")
plt.show()
```

```python
# Create a complete graph K6
K6 = nx.complete_graph(6)

# Define node positions for a symmetric layout
pos = nx.spring_layout(K6, seed=42)

# Split edges: all red except one blue edge
edges = list(K6.edges())
blue_edge = [edges[0]]  # Choose the first edge to be blue
red_edges = edges[0:]  # All other edges are red

# Plot the graph
plt.figure(figsize=(8, 8))
nx.draw(
    K6,
    pos,
    with_labels=True,
    node_color="lightgray",
    node_size=800,
    font_size=12,
    font_color="black",
    linewidths=1,
    edge_color="gray"  # Default edge color for any edges not specified
)

# Draw red edges
nx.draw_networkx_edges(
    K6,
    pos,
    edgelist=red_edges,
    edge_color="red",
    width=2
)

# Draw the blue edge
nx.draw_networkx_edges(
    K6,
    pos,
    edgelist=blue_edge,
    edge_color="blue",
    width=2
)

plt.title("Complete Graph K5 with Red Edges and One Blue Edge")
plt.show()
```
